```python
# 1から10000までの乱数を5000桁生成

a = np.random.randint(1, 10000, 5000)

# バブルソートの時間計測
import time
start = time.time()
bubble_sort.sort(a)
end = time.time()

print("bubble_sort: ", end - start)

# クイックソートの時間計測
start = time.time()
quick_sort.sort(a)
end = time.time()

print("quick_sort: ", end - start)

# 選択ソートの時間計測
start = time.time()
selection_sort.sort(a)
end = time.time()

print("selection_sort: ", end - start)

# 挿入ソートの時間計測
start = time.time()
insertion_sort.sort(a)
end = time.time()

print("insertion_sort: ", end - start)

# 基数ソートの時間計測
start = time.time()
radix_sort.sort(a)
end = time.time()

print("radix_sort: ", end - start)

# ヒープソートの時間計測
start = time.time()
heap_sort.sort(a)
end = time.time()

print("heap_sort: ", end - start)

# マージソートの時間計測
start = time.time()
merge_sort.sort(a)
end = time.time()

print("merge_sort: ", end - start)
```

```python
# xは1000刻み、1000から100000まで
x = [i for i in range(1000, 10000, 500)]

bubble_time = []
quick_time = []
selection_time = []
insertion_time = []
radix_sort_time = []
heap_sort_time = []
merge_sort_time = []

for i in x:
    a = np.random.randint(1, 10000, i)
    start = time.time()
    bubble_sort.sort(a)
    end = time.time()
    bubble_time.append(end - start)

    start = time.time()
    quick_sort.sort(a)
    end = time.time()
    quick_time.append(end - start)

    start = time.time()
    selection_sort.sort(a)
    end = time.time()
    selection_time.append(end - start)

    start = time.time()
    insertion_sort.sort(a)
    end = time.time()
    insertion_time.append(end - start)

    start = time.time()
    radix_sort.sort(a)
    end = time.time()
    radix_sort_time.append(end - start)

    start = time.time()
    heap_sort.sort(a)
    end = time.time()
    heap_sort_time.append(end - start)

    start = time.time()
    merge_sort.sort(a)
    end = time.time()
    merge_sort_time.append(end - start)

import matplotlib.pyplot as plt

plt.plot(x, bubble_time, label="bubble_sort")
plt.plot(x, quick_time, label="quick_sort")
plt.plot(x, selection_time, label="selection_sort")
plt.plot(x, insertion_time, label="insertion_sort")
plt.plot(x, radix_sort_time, label="radix_sort")
plt.plot(x, heap_sort_time, label="heap_sort")
plt.plot(x, merge_sort_time, label="merge_sort")
plt.legend()
plt.show()
```
